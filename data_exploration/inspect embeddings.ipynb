```python
# save cluster names
for NUM_CLUSTERS in range(6, 11):
    means = np.load(OUTPUT_DIR + 'cluster_'+str(NUM_CLUSTERS)+'_means.npy')
    keywords_mapper = {i: ", ".join(neighbors_vector(m, glove).head(10).index) for i, m in enumerate(means)}
    with open(OUTPUT_DIR + 'kmeans_' + str(NUM_CLUSTERS) + '_words.json', 'w') as f:
        f.write(json.dumps(keywords_mapper))
```
